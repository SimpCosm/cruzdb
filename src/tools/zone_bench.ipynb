```python
%matplotlib inline
import pandas as pd
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12.0, 6.0)
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./set.csv")
df["iops"] = 10**9 // df["ns"]
```

```python
df2 = df.groupby(["zonelen", "nops", "expr"]).mean()
```

```python
fig, ax = plt.subplots(1,2)
df2.unstack([-1, -3]).ns.vec.plot.bar(grid=True, log=True, ax=ax[0])
df2.unstack([-1, -3]).iops.vec.plot.bar(grid=True, ax=ax[1])
ax[1].yaxis.set_major_formatter(FormatStrFormatter('%d'))
```

```python
fig, ax = plt.subplots(1,2)
df2.unstack([-1, -3]).ns.set.plot.bar(grid=True, ax=ax[0])
df2.unstack([-1, -3]).iops.set.plot.bar(grid=True, ax=ax[1])
ax[1].yaxis.set_major_formatter(FormatStrFormatter('%d'))
```
